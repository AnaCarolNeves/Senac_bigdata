```python
import  pandas as pd #trabalha com dados maiores
import numpy as np #trabalha com dados menores porém transforma em pequenos blocos (arrays) os grandes volumes de dados
import random
import matplotlib.pyplot as plt
```

```python

amostra=np.array([2500,3200,2700,3000,3400,3700,4500,4700,5000,5200,6000,7000]) 
amostra
```

```python

media=np.mean(amostra)
mediana=np.median(amostra)
print(f'Média: {media}')
print(f'Mediana: {mediana}')
```

```python

distancia=(media-mediana)/mediana*100
print(f'Distândia entre média e mediana: {distancia:.2f}%')
```

```python
q1=np.percentile(amostra,25)
q2=np.percentile(amostra,50)
q3=np.percentile(amostra,75)

#Resumo estatístico:

print(f'Média:{media}')
print(f'Mediana:{mediana}')
print(f'Distância entre média e mediana:{distancia:.2f}%') 
print(f'Primeiro Quartil (Q1): {q1}')
print(f'Primeiro Quartil (Q1): {q2}')
print(f'Primeiro Quartil (Q1): {q3}')

```

```python
#Intervalo interquartil

if q1<= media <=q3:
    print("A média está dentro do intervalo interquartil, o que sugere confiabilidade")
else:
    print("A média está fora do intervalo interquartil, o que sugere menos confiabilidade")
```

```python
iqr=q3-q1
limite_superior=q3+(1.5*iqr)
limite_superior=q1+(1.5*iqr)
```

```python
array_r= np.random.randint(0,100,200) # os dois primeiros números são o intervalo que eu quero os números aleatórios (de 0  100), o último nº é a quantidade de nºs que eu quero. 
array_r
```

```python

amostra_df=pd.DataFrame(amostra, columns=['Valores'])
print(amostra_df)
```

```python
#Medidas de Dispersão
valor_maximo=amostra_df['Valores'].max()
valor_minimo=amostra_df['Valores'].min()
amplitude_total=valor_maximo-valor_minimo

print(f'Menor Valor:{valor_minimo}')
print(f'Maior Valor:{valor_maximo}')
print(f'Amplitude:{amplitude_total}')
```

```python
#configuração do painel
fig,ax=plt.subplots(1,2,figsize=(12,6)) #subplots são os plots que eu quero criar, nesse caso o plot 1 e o 2, lado a lado [ | ].

#Boxplot
ax[0].boxplot(amostra_df['Valores'])
ax[0].set_title('Distribuição de Dados - Boxplot')
ax[0].set_ylabel('Valores')

#Exibindo as Medidas
ax[1].text(0.1,0.8,f"Menor Valor:{valor_minimo}")
ax[1].text(0.1,0.7,f"Maior Valor:{valor_maximo}")
ax[1].text(0.1,0.6,f"Amplitude Total:{amplitude_total}")
ax[1].set_axis_off()

plt.tight_layout()
plt.show()



```
